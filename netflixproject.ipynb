```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
```

```python
netflix=pd.read_csv('netflix_sample_data.csv')
netflix.info()
netflix
```

```python
print(netflix.isnull().sum())
```

```python
netflix.dropna(subset=['title','type'] , inplace=True)
print(netflix.isnull().sum())
```

```python
netflix['country'].fillna('Unknown' , inplace=True)
netflix['director'].fillna('Not Mentioned' , inplace=True)
netflix
```

```python
sns.countplot(data=netflix, x='type')
plt.title("Total count of movies and Tv shows")
```

```python
count_country=netflix['country'].value_counts().head(10)
```

```python
sns.barplot(x=count_country.values , y=count_country.index , color='blue')
plt.title("Top 10 Countries by Number of Titles")
plt.xlabel("Number of Titles")
plt.show()
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix['year_added'] = netflix['date_added'].dt.year

sns.countplot(data=netflix, x='year_added', hue='type', palette='cool')
plt.title("Titles Added per Year")
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title='Type')
plt.show()

netflix
```

```python
genre_list=netflix['listed_in'].dropna().str.split(' , ')
genre_list
```

```python
all_genres = [genre for sublist in genre_list for genre in sublist]
all_genres
```

```python
movies_netflix=netflix[netflix['type']=='Movie']
movies_netflix['duration_int']=movies_netflix['duration'].str.extract('(\d+)').astype(float)
movies_netflix
```

```python
sns.histplot(movies_netflix['duration_int'], kde=True, bins=30, color='skyblue')
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(data=netflix, y='rating', order=netflix['rating'].value_counts().index, palette='cubehelix')
plt.title("Distribution of Ratings")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
long_movies = movies_netflix[movies_netflix['duration_int'] > 100]
print("Number of movies > 100 minutes:", len(long_movies))
```
